# Home Run Derby Analysis

This project analyzed real home run derby data from [https://baseballsavant.mlb.com/hr_derby].  It looks at all of the exit velocities with their corresponding diatances travelled.  The project then looks at whether or not a pattern exists between an ideal launch angle and a maximum distance.  Knowing this will allow for more home runs to be hit.

# Research Question
Do the furtherst home run hits, at a given speed, correlate to a similar launch angle?

# Assumptions
Assuming all baseball hits have same spin, backwards and sideways, as no data was provided by [mlb.com].

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from scipy.optimize import curve_fit
```

```python
df = pd.read_csv('baseballdata.txt', sep='\t')
```

```python
vel = np.array(df['Exit Velocity (MPH)'])
vel=vel.astype(int)
dist = np.array(df['Distance (Ft)'])
theta = np.array(df['Launch Angle'])
```

This is a plot of all the launch velocities vs. their corresponding distances.

```python
plt.plot(vel, dist, ".")
plt.title("Distance vs. Exit Velocity (MPH)")
plt.xlabel('Exit Velocity (MPH)')
plt.ylabel('Distance (Ft)')
```

There is a clear seperation of data on the graph.  Just about all of the points lie below a linear line, except for two outliers.  Taking a closer look at the data along the edge shows the following:

```python
#group all distances by speed
groups = {}
speeds = np.unique(vel)
maxdist = []
LAs = {}

for s in speeds:
    groups[s] = [] #store distances in a list
    
for i in range(len(vel)):
    v = vel[i]
    d = dist[i]
    groups[v].append(d)

for v in speeds:
    maxdist.append(np.max(groups[v]))

for i in range(len(speeds)):
    s = speeds[i]
    maxd = maxdist[i]
    for i in range(len(vel)): #loop through all data
        v = vel[i]
        d = dist[i]
        if v == s and d==maxd:
            LAs[maxd] = theta[i]

#Linear fit - regression line
def LinearFit(P,m,b):
    return m*P+b
bestParams,covar = curve_fit(LinearFit, speeds, maxdist)
#Variance if needed
# bestParams, variance = curve_fit(LinearFit,speeds,angles)
ydataFit = LinearFit(speeds, bestParams[0], bestParams[1])

plt.plot(speeds,maxdist, ".")
plt.plot(speeds,ydataFit, "r-")
plt.title("Max Distance vs Speed")
plt.ylabel('Distance (Ft)')
plt.xlabel('Exit Velocity (MPH)')
```

This plot shows the minimum speed needed to reach the corresponding distances to make a home run.  After further investigation, there is a clear, linear relationship between these hits.  The question arises: is there a relationship between their corresponding launch angles?

```python
angles = []
for d in maxdist:
    angles.append(LAs[d])

#Linear fit - regression line
def LinearFit2(P,m,b):
    return m*P+b
bestParams,covar = curve_fit(LinearFit, speeds, angles)
ydataFit2 = LinearFit2(speeds, bestParams[0], bestParams[1])
```

```python
plt.plot(speeds,angles, ".")
plt.plot(speeds,ydataFit2, "r-")
plt.title("Launch Angle of Max Distance Reached vs Speed")
plt.xlabel('Exit Velocity (MPH)')
plt.ylabel('Launch Angle (Degrees)')
```

# Conclusion
After closer examinaition, the ideal launch angle to get the furthest baseball hit falls between 25 to 35 degrees.  This correlation is not ideal, as 10 degrees variation is a lot when all possible launch angles range from 0-90 degrees.  In an ideal world, the previous plot would have a horizontal regression line, showing a consistent launch angle no matter what the launch speed was.
